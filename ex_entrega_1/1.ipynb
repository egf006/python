## 1) Algoritmo de ordenação

O **Bubble Sort** é um algoritmo de ordenação simples. A ideia central é:

> Percorrer a lista repetidamente, comparando pares de elementos adjacentes e trocando-os se estiverem na ordem errada.

---

### Intuição

Pense nos maiores elementos como “bolhas” que sobem para o final da lista a cada passagem.

---

### Como funciona (passo a passo)

1. Compare o elemento atual com o próximo  
2. Se estiverem fora de ordem, troque-os  
3. Avance para o próximo par  
4. Ao final de uma passada, o maior elemento estará na última posição  
5. Repita o processo ignorando os elementos já ordenados no final  
6. Pare quando não houver mais trocas  

### Exemplo

Lista inicial: [5, 3, 8, 4]

1ª passada:

- 5 > 3 → troca → [3, 5, 8, 4]

- 5 < 8 → ok

- 8 > 4 → troca → [3, 5, 4, 8] (8 já está no lugar certo)

2ª passada:

- 3 < 5 → ok

- 5 > 4 → troca → [3, 4, 5, 8]

3ª passada:

- 3 < 4 → ok → nenhuma troca → algoritmo pode parar

Implemente uma função que receba uma lista com valores numéricos e retorne uma lista com os elementos ordenados. Utilize o algoritmo *bubble sort* para fazer a ordenação.

Um investimento possui os seguintes fluxos de caixa:

- \( t = 0 \): investimento inicial de **-1000**
- \( t = 1 \): entrada de **300**

A **Taxa Interna de Retorno (TIR)** é a taxa \( r \) que zera o valor presente líquido (VPL):

$$
VPL(r) = -1000 + \frac{300}{(1+r)} + \frac{400}{(1+r)^2} + \frac{500}{(1+r)^3} = 0
$$

$$
