# Introdução ao Python para processamento numérico e análise de dados

# Pandas

O Pandas é uma biblioteca usada para análise de dados em Python.

## 1. Carregando Dados

É possível carregar dados de diversos formatos, por exemplo, o CSV. Os dados são carregados em objetos do tipo DataFrame:

````python
import pandas as pd

df_carregado = pd.read_csv('pessoas.csv')

print("\nPrimeiras 5 linhas do df_carregado:")
df_carregado.head()
````

## 2. Seleção de Colunas

Selecionando uma coluna do DataFrame:

````python
### Seleção de uma única coluna (resulta em Series):
coluna_nome = df_carregado['Nome']
print("Tipo da coluna 'Nome':", type(coluna_nome))
print("Primeiras 5 linhas da coluna 'Nome':\n", coluna_nome.head())
````

Também é possível selecionar múltiplas colunas:

````python
### Seleção de múltiplas colunas (resulta em DataFrame):
colunas_nome_idade = df_carregado[['Nome', 'Idade']]
print("Tipo das colunas ['Nome', 'Idade']:", type(colunas_nome_idade))
print("Primeiras 5 linhas das colunas 'Nome' e 'Idade':\n", colunas_nome_idade.head())
````

## 3. Criação de Novas Colunas

Criando uma coluna com valores `None`:

````python
df_carregado['Departamento'] = None
df_carregado.columns
````

````python
df_carregado.head()
````

Criando uma coluna calculada a partir de outra coluna:

````python
df_carregado['Salario_Anual'] = df_carregado['Salario'] * 12

df_carregado.head()
````

## 4. Filtragem de Linhas

Filtrando linhas com base em uma condição:

````python
### Filtragem com uma única condição (Idade > 25):
df_idade_maior_25 = df_carregado.loc[df_carregado['Idade'] > 25]

df_idade_maior_25.head()
````

Filtrando com base em múltiplas condições:

````python
### Filtragem com múltiplas condições (Cidade == 'São Paulo' E Salario > 5000):
df_multi_condicao = df_carregado.loc[(df_carregado['Cidade'] == 'São Paulo') & (df_carregado['Salario'] >= 5000)]

df_multi_condicao.head()
````

## 5. Seleção de Linhas Específicas

É possível selecionar linhas usando rótulos ou índices:

*   **`loc` (Label-based indexing)**: Seleciona linhas e colunas por seus rótulos (labels) ou um array booleano. Se o índice do DataFrame for numérico e padrão (começando de 0), `loc` pode se comportar de forma semelhante a `iloc` para seleção de linhas por número, mas sua intenção é sempre buscar pelo *rótulo* do índice.
*   **`iloc` (Integer-location based indexing)**: Seleciona linhas e colunas por sua posição inteira. É estritamente baseado na posição numérica (0-indexed) ao longo do eixo, independentemente do rótulo do índice do DataFrame.

Exemplo de seleção com rótulo:

````python
### Seleção de uma linha específica usando .loc (por rótulo/índice):
linha_loc_0 = df_carregado.loc[0]

print("Linha com índice 0 (usando .loc):\n", linha_loc_0)
````

Exemplo de seleção usando índice:

````python
### Seleção de uma linha específica usando .iloc (por posição inteira):
linha_iloc_0 = df_carregado.iloc[0]

print("Linha na posição 0 (usando .iloc):\n", linha_iloc_0)
````

Observação: Para DataFrames com índices numéricos padrão (0, 1, 2, ...), `.loc[0]` e `.iloc[0]` podem retornar a mesma linha, mas .loc busca pelo rótulo do índice e .iloc pela posição do índice. Para esclarecer, vamos considerar um dataframe com um índice que não é numérico:

````python
df_carregado_2 = df_carregado.set_index("Nome")
df_carregado_2
````

````python
df_carregado_2.loc["Bob"]
````

````python
df_carregado_2.iloc[1]
````

## 6. Ordenação de Dados

É possível ordenar os dados de um DataFrame com base nos valores de uma ou mais colunas.

````python
### Ordenação por 'Idade' (crescente):"
df_ordenado_idade = df_carregado.sort_values(by='Idade', ascending=True)
print("DataFrame ordenado por Idade (crescente):\n", df_ordenado_idade)
````

````python
### Ordenação por 'Salario' (decrescente):"
df_ordenado_salario = df_carregado.sort_values(by='Salario', ascending=False)
print("DataFrame ordenado por Salario (decrescente):\n", df_ordenado_salario)
````

````python
### Ordenação por 'Cidade' (crescente) e 'Idade' (decrescente):"
df_ordenado_multi = df_carregado.sort_values(by=['Cidade', 'Idade'], ascending=[True, False])
print("DataFrame ordenado por Cidade e Idade:", df_ordenado_multi)
````

## 7. Estatísticas Descritivas

É possível obter estatísticas sobre os dados das colunas com os métodos`describe()`, `mean()`, `median()`, `min()`, `max()`, e `value_counts()`.

````python
### Estatísticas descritivas para a coluna 'Salario' (utilizando .describe()):
df_carregado['Salario'].describe()
````

````python
### Média da coluna 'Idade':
df_carregado['Idade'].mean()
````

````python
### Mediana da coluna 'Salario':
df_carregado['Salario'].median()
````

````python
### Mínimo e Máximo da coluna 'Idade':
min_idade = df_carregado['Idade'].min()
max_idade = df_carregado['Idade'].max()
print(f"Idade Mínima: {min_idade}")
print(f"Idade Máxima: {max_idade}")
````

````python
### Contagem de valores únicos para a coluna 'Cidade' (utilizando .value_counts()):"
df_carregado['Cidade'].value_counts()
````

## 8. Renomear Colunas

É possível renomear colunas com o método `rename()`:

````python
### Renomeando a coluna 'Salario' para 'Renda Mensal':
df_carregado_renomeado = df_carregado.rename(columns={'Salario': 'Renda Mensal'})
print("Primeiras 5 linhas do DataFrame com a coluna renomeada:")
df_carregado_renomeado.head()
````

Também é possível renomear mais de uma coluna de uma só vez:

````python
### Renomeando as colunas 'Nome' para 'Primeiro Nome' e 'Cidade' para 'Localidade':
df_carregado_multi_renomeado = df_carregado.rename(columns={'Nome': 'Primeiro Nome', 'Cidade': 'Localidade'})

print("Primeiras 5 linhas do DataFrame com múltiplas colunas renomeadas:")
df_carregado_multi_renomeado.head()
````

## 9. Agregações com Groupby

O `groupby` permite fazer agregações (soma, média, etc.) em grupos de dados.

````python
### Média da Idade por Cidade:
df_carregado.groupby('Cidade')['Idade'].mean()
````

Também é possível agregar mais de uma coluna por vez, assim como fazer mais de um tipo de agregação (soma e média, por exemplo):

````python
### Soma e Média de Salário e Salário Anual por Cidade:\n"

df_carregado.groupby('Cidade')[['Salario', 'Salario_Anual']].agg(['sum', 'mean'])
````

## 10. Iteração sobre Linhas

Caso necessário, é possível iterar ao longo das linhas de um DataFrame usando `iterrows()`:

````python
### Iterando sobre as linhas do DataFrame usando iterrows():

for index, row in df_carregado.iterrows():
    print(f"Índice: {index}, Nome: {row['Nome']}, Cidade: {row['Cidade']}")
````

### Vantagens e Desvantagens de `iterrows()`

**Vantagens:**
*   **Acesso fácil a `index` e `row`**: `iterrows()` retorna o índice de cada linha e a linha como uma Series, facilitando o acesso a ambos.
*   **Leitura clara**: Para iterações simples onde o contexto de cada linha é importante, o código pode ser mais legível.

**Desvantagens:**
*   **Lento para DataFrames grandes**: `iterrows()` é significativamente mais lento do que métodos vetorizados, como a criação de colunas calculadas a partir de outras, mostrada anteriormente;
*   **Modificações não seguras**: A documentação do Pandas adverte explicitamente que `iterrows()` não deve ser usado para modificar DataFrames. As modificações em `row` não são refletidas no DataFrame original, e criar um novo DataFrame linha por linha é ineficiente. Para modificações, é preferível usar métodos como `loc` ou `iloc`.

## 11. Salvando DataFrames

É possível salvar dataframes em diversos formatos. Dentre os mais utilizados, estão o CSV e o pickle:

````python
### Salvando DataFrame para CSV:
df_carregado.to_csv('dados_salvos.csv', index=False)
print("DataFrame salvo como 'dados_salvos.csv' com sucesso.")
````

````python
### Salvando DataFrame para Pickle:
df_carregado.to_pickle('dados_salvos.pkl')
print("DataFrame salvo como 'dados_salvos.pkl' com sucesso.")
````

## 12. Conversão para Array NumPy

Muitas vezes é necessário converter os dados para o formato NumPy, para utilizar funções externas que esperam esse formato. Para isso, pode-se usar o método `to_numpy()`:

````python
### Conversão do DataFrame para Array NumPy (utilizando .to_numpy()):
np_array_dados = df_carregado.to_numpy()

print("Tipo da variável np_array_dados:", type(np_array_dados))
print("\nConteúdo do Array NumPy:\n", np_array_dados)
````

Em geral, converte-se apenas dados numéricos:

````python
df_carregado["Salario"].to_numpy()
````
