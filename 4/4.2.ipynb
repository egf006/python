# Visualização de dados com Seaborn

O Seaborn é uma biblioteca de visualização de dados Python baseada no Matplotlib que fornece uma interface de alto nível para desenhar gráficos estatísticos atraentes e informativos.

Por que Seaborn?

*   **Estética Aprimorada**: Os gráficos do Seaborn são visualmente mais agradáveis por padrão.
*   **Fácil de Usar**: Simplifica a criação de gráficos complexos com poucas linhas de código.
*   **Estatísticas Integradas**: Facilita a visualização de relações e distribuições de dados estatísticos.
*   **Integração com Pandas**: Funciona perfeitamente com DataFrames do Pandas.

Vamos considerar um conjunto de dados de preços de acões de empresas:

```python
import pandas as pd

df = pd.read_csv("empresas.csv")
df.head()
```

Antes de fazer os gráficos com o Seaborn, vamos importar algumas bibliotecas e fazer algumas configurações de visualização:

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Configurações para melhorar a visualização dos gráficos
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. `sns.lineplot()`: Visualizando Tendências ao Longo do Tempo

O `lineplot` é ideal para mostrar como uma variável numérica muda ao longo do tempo ou de uma sequência ordenada. É excelente para séries temporais como preços de ações.

```python
plt.figure(figsize=(12, 7))
sns.lineplot(data=df, x='Date', y='Price', hue='Company', marker='o', lw=1.5)
plt.title('Preço das Ações ao Longo do Tempo por Empresa')
plt.xlabel('Data')
plt.ylabel('Preço das Ações')
plt.xticks(rotation=45)
plt.legend(title='Empresa')
plt.tight_layout()
plt.show()
```

Para tornar a visualização mais limpa, é possível filtrar por data:

```python
# Verificando o tipo da coluna Date
df.dtypes
```

```python
# Convertendo para datetime
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
# Exemplo de filtro
df.loc[df['Date'] > '2022-04-01']
```

```python
# Plot com filtro
plt.figure(figsize=(12, 7))
sns.lineplot(data=df.loc[df['Date'] > '2022-04-01'], x='Date', y='Price', hue='Company', marker='o', lw=1.5)
plt.title('Preço das Ações ao Longo do Tempo por Empresa')
plt.xlabel('Data')
plt.ylabel('Preço das Ações')
plt.xticks(rotation=45)
plt.legend(title='Empresa')
plt.tight_layout()
plt.show()
```

## 2. `sns.barplot()`: Comparando Categorias

O `barplot` é usado para representar a estimativa pontual e o intervalo de confiança de uma variável numérica para cada categoria de uma variável qualitativa. É ótimo para comparar médias ou somas entre grupos.

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Sector', y='MarketCap', estimator=np.mean, errorbar='sd')
plt.title('Capitalização de Mercado Média por Setor')
plt.xlabel('Setor')
plt.ylabel('Capitalização de Mercado Média ($)')
plt.tight_layout()
plt.show()
```

`errorbar` pode ser configurado como (ver https://seaborn.pydata.org/tutorial/error_bars.html):
- `sd` - standard deviation
- `se` - standard error
- `pi` - percentile interval
- `ci` - confidence interval

## 3. `sns.scatterplot()`: Relação entre Duas Variáveis Numéricas

O `scatterplot` é ideal para visualizar a relação entre duas variáveis numéricas. Ele pode ajudar a identificar correlações, aglomerados ou padrões. Podemos adicionar uma terceira variável para codificar cor ou tamanho.

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='PE_Ratio', y='MarketCap', hue='Sector', size='Volume', sizes=(50, 500), alpha=0.7, palette='Spectral')
plt.title('Relação entre PE_Ratio e Capitalização de Mercado, colorida por Setor e dimensionada por Volume')
plt.xlabel('Índice P/L (PE_Ratio)')
plt.ylabel('Capitalização de Mercado')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.tight_layout()
plt.show()
```

## 4. `sns.boxplot()`: Distribuição e Outliers em Categorias

O `boxplot` (ou gráfico de caixa) é uma excelente ferramenta para exibir a distribuição de dados numéricos e identificar valores atípicos (*outliers*) para diferentes grupos. Ele mostra a mediana, quartis e a dispersão dos dados. (Ref. https://pt.wikipedia.org/wiki/Diagrama_de_caixa).

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Company', y='Price')
plt.title('Distribuição dos Preços das Ações por Empresa')
plt.xlabel('Empresa')
plt.ylabel('Preço das Ações')
plt.tight_layout()
plt.show()
```
