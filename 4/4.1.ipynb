# Introdução ao Python para processamento numérico e análise de dados

# Matplotlib

O Matplotlib é a biblioteca mais utilizada para construção de gráficos matemáticos e visualizações de dados com o Python.

## 1. Gráfico XY a partir de dados em arrays NumPy

Primeiramente, vamos gerar alguns dados:

```python
import numpy as np

# Gerando dados de exemplo
x = np.linspace(0, 10, 100) # 100 pontos entre 0 e 10
y = np.sin(x) # Valores de seno correspondentes a x
```

Podemos fazer um gráfico XY simples a partir desses dados com:

```python
import matplotlib.pyplot as plt

plt.plot(x, y)

# Exibindo o gráfico
plt.show()
```

O gráfico é composto por uma sequência de pontos (x, y) obtidos dos respectivos arrays NumPy, conectados por linhas retas.

## 2. Personalização de Linhas

É possível personalizar a aparência das linhas no seu gráfico, alterando a cor, o estilo e a largura.

Para alterar a cor, você pode usar um nome de cor (ex: `'red'`, `'green'`, `'blue'`) ou um código hexadecimal. Para o estilo da linha, existem opções como `'--'` (tracejada), `'-.'` (traço-ponto), `':'` (pontilhada) e `'-'` (sólida, padrão).

Vamos modificar o plot anterior para ter uma linha vermelha e tracejada.

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando os mesmos dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Criando o plot com personalização de linha
# 'r--' significa linha vermelha tracejada
plt.plot(x, y, 'r--')

# Você também pode especificar as propriedades individualmente:
#plt.plot(x, y, color='red', linestyle='--')

# Exibindo o plot
plt.show()
```

## 3. Adicionando Rótulos e Título

É possível adicionar rótulos para os eixos X e Y, um título para o gráfico e, se tiver várias séries de dados, uma legenda para identificá-las.

- `plt.xlabel()`: Define o rótulo do eixo X.
- `plt.ylabel()`: Define o rótulo do eixo Y.
- `plt.title()`: Define o título do gráfico.
- `plt.legend()`: Exibe a legenda. Para que funcione, cada `plt.plot()` deve ter um argumento `label` especificado.

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando os mesmos dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Criando o plot com personalização de linha e um label para a legenda
plt.plot(x, y, color='blue', linestyle='-', label='Seno')

# Adicionando rótulos aos eixos
plt.xlabel('t (s)')
plt.ylabel('y(t)')

# Adicionando um título ao gráfico
plt.title('Gráfico da Função Seno')

# Adicionando a legenda (necessário 'label' no plt.plot para funcionar)
plt.legend()

# Exibindo o plot
plt.show()
```

## 4. Múltiplos Plots

Para plotar várias curvas em uma mesma figura, basta repetir o comando `plt.plot()` para cada uma das curvas:

```python
# Gerando dados de exemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Criando uma figura com dois subplots
#plt.figure(figsize=(10, 5)) # Opcional: define o tamanho geral da figura

# Primeiro subplot
plt.plot(x, y1, color='red', label='Seno')
plt.plot(x, y2, color='green', label='Cosseno')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

Também é possível compor vários gráficos em uma mesma figura com a função `plt.subplot()`, que divide a área da figura em uma grade e seleciona uma área específica para o plot atual.

A função `plt.subplot(nrows, ncols, index)` espera três argumentos:
- `nrows`: O número de linhas da grade.
- `ncols`: O número de colunas da grade.
- `index`: O índice do subplot atual (começando em 1, da esquerda para a direita e de cima para baixo).

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando dados de exemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Primeiro subplot
plt.subplot(1, 2, 1) # 1 linha, 2 colunas, primeiro subplot
plt.plot(x, y1, color='red', label='Seno')
plt.title('Função Seno')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Segundo subplot
plt.subplot(1, 2, 2) # 1 linha, 2 colunas, segundo subplot
plt.plot(x, y2, color='green', linestyle='--', label='Cosseno')
plt.title('Função Cosseno')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.tight_layout() # Ajusta automaticamente os parâmetros do subplot para que caibam na área da figura
plt.show()
```

## 5. Tamanho da Figura

O tamanho padrão de uma figura no Matplotlib pode não ser o ideal para todas as suas necessidades de visualização. Você pode controlar as dimensões da sua figura usando o argumento `figsize` na função `plt.figure()`.

`figsize` aceita uma tupla `(largura, altura)` em polegadas. Vamos criar um plot com um tamanho de figura personalizado.

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando dados de exemplo
x = np.linspace(0, 10, 100)
y = np.cos(x)

# Criando uma figura com um tamanho personalizado (12 polegadas de largura por 4 de altura)
plt.figure(figsize=(12, 4))

# Criando o plot
plt.plot(x, y, color='purple', linestyle=':', label='Cosseno de X')

# Adicionando rótulos e título
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Função Cosseno com Tamanho Personalizado')
plt.legend()

# Exibindo o plot
plt.show()
```
