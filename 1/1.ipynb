## 9. Iteráveis e laço for pythônico

A função `range` costuma ser usada em conjunto com o laço for. `range(x)` retorna um iterador com valores de `0` a `x-1`:

````python
range(10)
````

O laço for faz o processo de iteração no objeto do tipo range:

````python
for i in range(10):
    print(i)
````

Também é possível iterar o objeto e armazenar os valores em uma lista com a função `list()`:

````python
list(range(10))
````

Suponha que você queira iterar sobre os elementos de uma lista. Por exemplo:

````python
nomes = ["Carlos", "João", "Maria", "Gabriel"]
````

Geralmente, em outras linguagens de programação, utiliza-se um laço for fazendo iterações em uma variável de controle, que é usada para carregar os dados da lista, como no seguinte trecho:

````python
for i in range(len(nomes)):
    print(nomes[i])
````

Em Python, essa não é uma prática comum. A forma mais pythônica de fazer a iteração em um laço for é utilizar diretamente a lista:

````python
for nome in nomes:
    print(nome)
````

Entretanto, fazendo dessa forma, não temos acesso aos índices `i` para cada elemento. Para isso, pode-se usar a função `enumerate()`:

````python
for i, nome in enumerate(nomes):
    print(f"{i} - {nome}")
````

## 10. List comprehensions

É uma forma compacta de construir listas.

Supondo que queremos construir uma lista com os quadrados dos números de 1 a 4:

````python
squares = []

for x in range(5):
    squares.append(x**2)

print(squares)
````

É possível fazer forma menos verbosa usando list comprehensions:

```python
[expression for item in iterable]
```

````python
squares = [x**2 for x in range(5)]
print(squares)
````

É possível usar condicionais:

```python
[expression for item in iterable if condition]
```


````python
pares = []

for x in range(10):
    if x % 2 == 0:
        pares.append(x)

print(pares)
````

````python
pares = [x for x in range(10) if x % 2 == 0]
print(pares)
````

## 11. Dicionários

## 12. Imports e namespaces

Python possui muitas funcionalidades prontas organizadas em módulos.

Para usar um módulo, precisamos importá-lo:

````python
import math

print(math.sqrt(9))
````

Aqui:
- `math` é o módulo
- `sqrt` é uma função dentro dele

Um namespace é um “espaço de nomes”.

Ele organiza nomes (variáveis, funções, classes) para evitar conflitos.

Quando fazemos:
```python
import math
```
Estamos dizendo “Coloque todas as funções dentro do espaço math”.

Por isso usamos:
```python
math.sqrt(9)
```
O `math.` indica onde procurar a função.


Para importar algo específico:

````python
from math import sqrt

print(sqrt(9))
````

⚠️ Mas cuidado: dessa forma, podem ser gerarados conflitos de nomes.

É comum usar apelidos:

````python
import math as m

print(m.sqrt(16))
````

Comum em bibliotecas como:

````python
import numpy as np
import pandas as pd
````

Namespaces são importantes.

Suponha por exemplo::

```
from math import *
```

- Isso importa tudo diretamente, no namespace padrão.

- Problema: Pode sobrescrever nomes existentes

- Deixa o código menos claro

Por isso, geralmente preferimos:

```python
import modulo
```

ou

```python
import modulo as apelido
```

## 15. Detalhes sobre variáveis em memória

Podemos pensar que Python associa tags às variáveis em memória:

````python
a = [0, 0, 0]
print(a)
````

````python
b = a
print(b)
````

Com isso, `a` e `b` apontam para o mesmo elemento em memória:

````python
a[1] = 1
print(a)
````

````python
print(b)
````

Para confirmar que `a` e `b` se referem ao mesmo elemento de memória, podemos usar a função `id()`:

````python
id(a)
````

````python
id(b)
````

Como podemos ver, são idênticos, confirmando que se referem ao mesmo elemento em memória.

Alguns objetos disponibilizam o método `copy()` para permitir a sua cópia em um elemento diferente na memória. A lista é um deles:

````python
b = a.copy()
````

````python
id(b)
````

````python
a[0] = 1
print(a)
````

````python
print(b)
````

Para objetos que não disponibilizam o método `copy()`, podemos usar a biblioteca `copy`:

````python
import copy
````

````python
A = [[1,2,3], [4,5,6]]
print(A)
````

````python
print(A[0])
````

Cópia rasa (shallow):

````python
B = copy.copy(A)
````

````python
id(A)
````

````python
id(B)
````

Como podemos notar, são objetos diferentes em memória. No entanto, se olharmos os elementos da lista:

````python
id(A[0])
````

````python
id(B[0])
````

É possível notar que se tratam dos mesmos objetos. Isso ocorre porque a cópia feita foi do tipo rasa (shallow). Para copiar todos os elementos internos, é necessário fazer uma cópia profunda (deep):

````python
C = copy.deepcopy(A)
````

````python
id(C)
````

````python
id(A)
````

````python
id(C[0])
````

````python
id(A[0])
````

## 16. Lendo e escrevendo arquivos

Em Python, usamos a função `open()`:

```
arquivo = open("dados.txt", "r")
```

O segundo parâmetro é o modo:
- `r` → leitura (read)
- `w` → escrita (write) – sobrescreve
- `a` → adicionar (append)
- `b` → modo binário

````python
arquivo = open("dados.txt", "r")
conteudo = arquivo.read()
print(conteudo)
arquivo.close()
````

Uma forma recomendada é usando o bloco `with`:

````python
with open("dados.txt", "r") as arquivo:
    conteudo = arquivo.read()
    print(conteudo)
````

Para escrever um arquivo:

````python
with open("saida.txt", "w") as arquivo:
    arquivo.write("Olá, mundo!\n")
````

## 17. Serialização com arquivos pickle

O módulo `pickle` permite serializar e desserializar objetos Python.

Serializar significa converter um objeto Python em uma sequência de bytes para salvar em arquivo.

Desserializar significa reconstruir o objeto original a partir do arquivo.

Isso permite salvar:
- Listas
- Dicionários
- Modelos treinados
- Objetos complexos

Para salvar um objeto (dump):

````python
import pickle

dados = {"nome": "Ana", "idade": 20, "notas": [8, 9, 10]}

with open("dados.pkl", "wb") as arquivo:
    pickle.dump(dados, arquivo)
````

Para ler um arquivo pickle:

````python
with open("dados.pkl", "rb") as arquivo:
    dados_carregados = pickle.load(arquivo)

print(dados_carregados)
````
